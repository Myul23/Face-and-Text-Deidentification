```python
import os
from shutil import move
```

### output 분리

```python
inputs = "texts"
outputs = "non_valid"
```

```python
for idx, f in enumerate(os.listdir(inputs)):
    if len(f.split("-")) > 2:
        input_path = os.path.join(inputs, f)
        output_path = os.path.join(outputs, f)
        move(input_path, output_path)
    if idx % 100 == 0:
        print(idx, "finish")
print("finished")
```

### output label 다르게 move

```python
inputs = "non_valid"
outputs = "texts"
valid = "valid"
images_path = "data"
```

```python
assert os.path.exists(inputs) and os.path.exists(images_path)

if not os.path.exists(outputs):
    os.mkdir(outputs)
if not os.path.exists(valid):
    os.mkdir(valid)
```

```python
for text_file in os.listdir(inputs):
    parts = text_file.split("-", 2)
    image_folder = os.path.join(images_path, parts[1])
    for image_file in os.listdir(image_folder):
        if parts[2].rsplit(".", 1)[0] == image_file.rsplit(".", 1)[0]:
            print(image_file)
            break
```

```python
count = 0
for text_file in os.listdir(inputs):
    parts = text_file.split("-", 2)
    try:
        flag = False
        for c in os.listdir(images_path):
            image_folder = os.path.join(images_path, c)
            for image_file in os.listdir(image_folder):
                if parts[2].rsplit(".", 1)[0] == image_file.rsplit(".", 1)[0]:
                    count += (parts[1] != c)
                    # save (move) image file
                    input_path = os.path.join(image_folder, image_file)
                    output_path = os.path.join(valid, parts[0] + "-" + image_file)
                    move(input_path, output_path)

                    # save text file
                    input_path = os.path.join(inputs, text_file)
                    output_path = os.path.join(outputs, parts[0] + "-" + parts[2])
                    move(input_path, output_path)

                    flag = True
                    break
            if flag:
                break
    except FileNotFoundError as error:
        pass
print(count)
```

### class number 바꾸기

```python
inputs = "valid data/texts"
```

```python
for f in os.listdir(inputs):
    path = os.path.join(inputs, f)
    with open(path, "r") as ff:
        lines = ff.readlines()

    text = ""
    for line in lines:
        text += "0 " + line.split(" ", 1)[1]

    with open(path, "w") as ff:
        ff.write(text)
```
